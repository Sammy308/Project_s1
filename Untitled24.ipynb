```python
#연도 이상 유무 확인
df2 = df.sort_values(by=['Year'], na_position='first')  # NaN이 먼저 뜨도록 설정, 확인 결과 NaN 값 271개 발견

df3 = df2[271:]  #NaN 값 외 데이터 따로 추출 16327개 + 판매량 데이터 중 표기 방식이 다른 자료 확인 : M,K 사용 → 변경 필요
df3
```

```python
df3 = df3.reset_index()  # 자꾸 반복하면 자꾸 index 추가됨

def cor_data(d):
  i = 0
  for v in d['Year']:
    if int(v) < 10:
      v = '200' + v
      d.loc[i,'Year'] = v
    elif int(v) < 20:
      v = '20' + v
      d.loc[i,'Year'] = v
    elif int(v) < 100:
      v = '19' + v
      d.loc[i,'Year'] = v
    elif int(v) > 100:
      pass
    i += 1
  return d

cor_data(df3)  # 한자리, 두자리로 적힌 곳에 앞자리 부여
df3['Year'] = df3['Year'].apply(int)
df3 = df3.set_index('index') # 연도수 수정 및 인덱스 재위치 완료
df3
```

```python
#온전히 숫자로 표기된 데이터 확인

# 북미지역 (숫자 = 16398개) ※ 위 M,K 데이터 수와 합치면 16598개 일치
df5_num = df4[~df4['NA_Sales'].str.contains('M|K', regex=True)]
df5_num = df5_num['NA_Sales']                   #숫자만 따로 추출

# 유럽지역 (숫자 = 16398개) ※ 위 M,K 데이터 수와 합치면 16598개 일치
df6_num = df4[~df4['EU_Sales'].str.contains('M|K', regex=True)]
df6_num = df6_num['EU_Sales']                   #숫자만 따로 추출

# 일본지역 (숫자 = 16399개) ※ 위 M,K 데이터 수와 합치면 16598개 일치
df7_num = df4[~df4['JP_Sales'].str.contains('M|K', regex=True)]
df7_num = df7_num['JP_Sales']                    #숫자만 따로 추출

# 기타지역 (숫자 = 16398개) ※ 위 M,K 데이터 수와 합치면 16598개 일치
df8_num = df4[~df4['Other_Sales'].str.contains('M|K', regex=True)]
df8_num = df8_num['Other_Sales']               #숫자만 따로 추출
```

```python
# M와 K 표기를 숫자로 변경 및 일 단위로 환산 (df5_: 미국, df6_: 유럽, df7_: 일본, df8_: 기타 지역)
#                                                            (기존 단위 : 백만 → 변경 단위 : 일)

def cal_mil(m):     # M 표기 데이터를 숫자로 계산하여 수정하는 함수
  m = m.str.replace('M','')
  m = m.reset_index()
  i = 0
  m.iloc[:,1] = m.iloc[:,1].astype(float)
  for o in m.iloc[:,1]:
    o = o*1000000
    m.iloc[i,1] = o
    i += 1
  m.iloc[:,1] = m.iloc[:,1].astype(int)
  m = m.set_index('index')
  return m

def cal_tho(k):     # K 표기 데이터를 숫자로 계산하여 수정하는 함수
  k = k.str.replace('K','')
  k = k.reset_index()
  i = 0
  k.iloc[:,1] = k.iloc[:,1].astype(float)
  for o in k.iloc[:,1]:
    o = o*1000
    k.iloc[i,1] = o
    i += 1
  k.iloc[:,1] = k.iloc[:,1].astype(int)
  k = k.set_index('index')
  return k
```

```python
# 숫자 데이터 일 단위로 환산 (df5_: 미국, df6_: 유럽, df7_: 일본, df8_: 기타 지역)
#                                     (기존 단위 : 백만 → 변경 단위 : 일)

def cal_num(n):  # 데이터 단위에 맞춰 계산하여 수정하는 함수
  n = n.reset_index()
  i = 0
  n.iloc[:,1] = n.iloc[:,1].astype(float)
  for o in n.iloc[:,1]:
    o = o*1000000
    n.iloc[i,1] = o
    i += 1
  n.iloc[:,1] = n.iloc[:,1].astype(int)
  n = n.set_index('index')
  return n
```
